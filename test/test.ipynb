```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df = pd.read_csv('temp.high.csv')
data = np.array(df)
prns = np.unique(data[:, 1])
```

```python
for plt_index in range(5, data.shape[1]-2):
    for prn in [5]:
        data_single = data[data[:, 1] == prn, :]
        data_single = data_single[np.argsort(data_single[:, 0])]
        # data_single = data_single[1::2, :]
        plt.plot(data_single[:, 0], data_single[:, plt_index], label = "SV:{sv}".format(sv = prn))
        plt.title(df.columns[plt_index])
    # plt.legend()
    plt.show()
```
