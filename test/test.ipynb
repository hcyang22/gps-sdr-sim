```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
path_highlevel = 'temp.high.csv'
path_lowlevel = 'temp.low.csv'
df = pd.read_csv(path_highlevel)
data = np.array(df)
prns = np.unique(data[:, 1])
```

```python
for plt_index in range(8, data.shape[1]-1):
    for prn in prns:
        data_single = data[data[:, 1] == 5, :]
        data_single = data_single[np.argsort(data_single[:, 0])]
        plt.plot(data_single[:, 0], data_single[:, plt_index])
        plt.title(df.columns[plt_index])
    plt.show()
```
